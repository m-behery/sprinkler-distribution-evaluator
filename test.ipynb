```python
import pandas as pd

df = pd.read_csv('./van 18.csv', header=None)
print(df.to_markdown())
```

```python
grid = pd.pivot(df, columns=0, index=1, values=2)
print(grid.to_markdown())
```

```python
table = grid.melt(ignore_index=False, var_name=0, value_name='tmp').reset_index().rename(columns={'tmp': 2})
table.sort_index(axis=1)
```
